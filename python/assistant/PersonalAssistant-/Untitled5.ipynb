```python
import sqlite3
conn = sqlite3.connect('XFQuestionsLib.db')
cursor = conn.cursor()
```

```python
sql = '''
SELECT * FROM Tests
ORDER BY ques_id

'''
cursor = cursor.execute(sql)
res = cursor.fetchall()
res
```

```python
sql = '''
SELECT ques_id,COUNT(*) FROM Tests
GROUP BY ques_id
ORDER BY COUNT(*)

'''
cursor = cursor.execute(sql)
res = cursor.fetchall()
res
```
